                              **THEORITICAL**

#1 What is a random variable in probability theory?
ANS: A random variable is a numerical value determined by the outcome of a random phenomenon. It assigns a real number to each possible outcome of a random experiment.

#2 What are the types of random variables?
ANS:

Discrete Random Variable: Takes countable values (like 0, 1, 2…)

Continuous Random Variable: Takes any value within a continuous range (like time, height, weight)

#3 What is the difference between discrete and continuous distributions?
ANS:

Discrete Distribution:

Probabilities are assigned to exact values

Example: Number of students in a class

Probability of an exact value: P(X = x)

Continuous Distribution:

Probabilities are assigned over intervals

Example: Weight of a person

Probability over a range: P(a ≤ X ≤ b)

#4 What are probability distribution functions (PDF)?
ANS: For continuous random variables, a PDF describes the likelihood of a random variable falling within a particular range of values.
For discrete variables, it’s called a probability mass function (PMF).

#5 How do cumulative distribution functions (CDF) differ from PDFs?
ANS:

CDF gives the probability that a random variable is less than or equal to a value x.

PDF (for continuous) or PMF (for discrete) gives the probability at an exact value.

Example:

CDF: P(X ≤ 3)

PDF: f(3)

#6 What is a discrete uniform distribution?
ANS: A distribution where each outcome has an equal probability.
Example: Rolling a fair die — P(1)=P(2)=…=P(6)=1/6.

#7 What are the key properties of a Bernoulli distribution?
ANS:

Two outcomes: success (1) or failure (0)

One trial

Probability of success = p

Expected value = p

Variance = p(1-p)

#8 What is the binomial distribution, and how is it used?
ANS:

Models the number of successes in n independent Bernoulli trials

Parameters: n (trials) and p (success probability)

Used for: Quality control, surveys, defective items in a batch

#9 What is the Poisson distribution and where is it applied?
ANS:

Models the number of events in a fixed interval of time or space

Events occur independently at a constant average rate λ (lambda)

Used for: Call arrivals, accidents per year, emails per hour

#10 What is a continuous uniform distribution?
ANS: A distribution where all values in a continuous range are equally likely.
Example: Randomly selecting a number between 0 and 1.

#11 What are the characteristics of a normal distribution?
ANS:

Symmetrical bell-shaped curve

Mean = Median = Mode

Defined by mean (μ) and standard deviation (σ)

68-95-99.7 Rule applies

#12 What is the standard normal distribution, and why is it important?
ANS:

A normal distribution with μ=0 and σ=1

Used to compute probabilities using Z-scores

Simplifies probability calculations and hypothesis testing

#13 What is the Central Limit Theorem (CLT), and why is it critical?
ANS:

States that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases

Important for inferential statistics, confidence intervals, and hypothesis testing

#14 How does the Central Limit Theorem relate to the normal distribution?
ANS: Because of the CLT, sample means of large samples follow a normal distribution, enabling use of normal probability techniques even for non-normal populations.

#15 What is the application of Z statistics in hypothesis testing?
ANS:

Used when population standard deviation is known

Compares sample statistic to population mean under normality (or by CLT)

Example: Testing if a machine fills bottles to a target volume

#16 How do you calculate a Z-score, and what does it represent?
ANS:
Formula:
Z = (X - μ) / σ
It represents how many standard deviations a value X is from the population mean μ.

#17 What are point estimates and interval estimates in statistics?
ANS:

Point estimate: Single value estimate of a parameter (e.g., sample mean)

Interval estimate: Range of values likely to contain the population parameter (e.g., confidence interval)

#18 What is the significance of confidence intervals in statistical analysis?
ANS: Confidence intervals provide a range within which the true population parameter is likely to lie, giving a measure of certainty (commonly 95%).

#19 What is the relationship between a Z-score and a confidence interval?
ANS:

Z-scores determine the margin of error for a confidence interval in a normal distribution.

Example: 95% CI uses a Z-score of 1.96

#20 How are Z-scores used to compare different distributions?
ANS: They standardize different distributions to the standard normal (mean = 0, SD = 1), allowing comparison on a common scale.

#21 What are the assumptions for applying the Central Limit Theorem?
ANS:

Independent and identically distributed (i.i.d) samples

Large enough sample size (typically n ≥ 30)

Finite variance and mean

#22 What is the concept of expected value in a probability distribution?
ANS: It’s the weighted average of all possible values a random variable can take, each weighted by its probability.

Formula:
E(X) = Σ P(x) × x

#23 How does a probability distribution relate to the expected outcome of a random variable?
ANS: A probability distribution describes all possible outcomes and their probabilities, from which the expected value (long-run average outcome) is calculated.

```python
#1 Generate a random variable and display its value
import random

value = random.randint(1, 100)
print("Random Variable Value:", value)
```

```python
#2 Generate a discrete uniform distribution and plot its PMF
import numpy as np
import matplotlib.pyplot as plt

values = np.arange(1, 7)
prob = np.ones_like(values) / len(values)

plt.stem(values, prob, basefmt=" ")
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Discrete Uniform Distribution (PMF)')
plt.show()
```

```python
#3 Bernoulli distribution PDF function
def bernoulli_pdf(x, p):
    return p if x == 1 else 1 - p

# Example
print("P(X=1):", bernoulli_pdf(1, 0.7))
print("P(X=0):", bernoulli_pdf(0, 0.7))
```

```python
#4 Simulate a binomial distribution (n=10, p=0.5) and plot histogram
from scipy.stats import binom

data = binom.rvs(n=10, p=0.5, size=1000)
plt.hist(data, bins=11, density=True)
plt.xlabel('Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.show()
```

```python
#5 Create a Poisson distribution and visualize it
from scipy.stats import poisson

x = np.arange(0, 15)
pmf = poisson.pmf(x, mu=4)

plt.stem(x, pmf, basefmt=" ")
plt.xlabel('Events')
plt.ylabel('Probability')
plt.title('Poisson Distribution (λ=4)')
plt.show()
```

```python
#6 CDF of a discrete uniform distribution
from scipy.stats import randint

x = np.arange(1, 7)
cdf = randint.cdf(x, 1, 7)

plt.step(x, cdf, where='mid')
plt.xlabel('Value')
plt.ylabel('CDF')
plt.title('CDF of Discrete Uniform Distribution')
plt.show()
```

```python
#7 Generate a continuous uniform distribution and visualize it
from scipy.stats import uniform

data = uniform.rvs(size=1000)
plt.hist(data, bins=20, density=True)
plt.title('Continuous Uniform Distribution')
plt.show()
```

```python
#8 Simulate normal distribution data and plot histogram
data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data, bins=30, density=True)
plt.title('Normal Distribution Histogram')
plt.show()
```

```python
#9 Calculate Z-scores and plot them
from scipy.stats import zscore

data = np.random.normal(size=100)
z_scores = zscore(data)

plt.hist(z_scores, bins=20)
plt.title('Z-Scores of Normal Data')
plt.show()
```

```python
#10 Central Limit Theorem simulation with non-normal distribution
samples = [np.mean(np.random.exponential(scale=2, size=30)) for _ in range(1000)]
plt.hist(samples, bins=30, density=True)
plt.title('CLT Applied to Exponential Distribution (Sample Means)')
plt.show()
```

```python
#11 Simulate multiple samples and verify CLT
sample_means = [np.mean(np.random.normal(size=50)) for _ in range(1000)]
plt.hist(sample_means, bins=30)
plt.title('Sample Means Approaching Normal (CLT)')
plt.show()
```

```python
#12 Standard normal distribution PDF
from scipy.stats import norm

x = np.linspace(-4, 4, 100)
pdf = norm.pdf(x)

plt.plot(x, pdf)
plt.title('Standard Normal Distribution (PDF)')
plt.show()
```

```python
#13 Generate random variables and binomial probabilities
data = binom.rvs(n=10, p=0.5, size=1000)
values, counts = np.unique(data, return_counts=True)
probabilities = counts / counts.sum()
for v, p in zip(values, probabilities):
    print(f"P(X={v}) = {p:.3f}")
```

```python
#14 Calculate Z-score for a given value
def calculate_z(x, mean, std):
    return (x - mean) / std

print("Z-score:", calculate_z(85, 70, 10))
```

```python
#15 Hypothesis testing using Z-statistic
import math

sample_mean = 70
population_mean = 75
population_std = 10
n = 30

z = (sample_mean - population_mean) / (population_std / math.sqrt(n))
print("Z-statistic:", z)
```

```python
#16 Confidence interval for a dataset
import scipy.stats as stats

data = np.random.normal(50, 10, 100)
mean = np.mean(data)
se = stats.sem(data)
conf_interval = stats.t.interval(0.95, len(data)-1, loc=mean, scale=se)
print("95% Confidence Interval:", conf_interval)
```

```python
#17 Confidence interval for normal data mean
data = np.random.normal(100, 15, 200)
mean = np.mean(data)
se = stats.sem(data)
conf_interval = stats.t.interval(0.95, len(data)-1, loc=mean, scale=se)
print("95% CI for mean:", conf_interval)
```

```python
#25 Write a Python program to compare the standard deviations of two datasets and visualize the difference

import numpy as np
import matplotlib.pyplot as plt

# Generate two datasets
data1 = np.random.normal(50, 5, 1000)
data2 = np.random.normal(50, 10, 1000)

# Calculate standard deviations
std1 = np.std(data1)
std2 = np.std(data2)

print("Standard Deviation of Dataset 1:", std1)
print("Standard Deviation of Dataset 2:", std2)

# Visualize with histograms
plt.hist(data1, bins=30, alpha=0.5, label='Std=5')
plt.hist(data2, bins=30, alpha=0.5, label='Std=10')
plt.title('Comparison of Standard Deviations')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
#26 Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution

# Generate normal data
data = np.random.normal(0, 1, 1000)

# Calculate Range
range_value = np.ptp(data)

# Calculate IQR (Interquartile Range)
q75, q25 = np.percentile(data, [75, 25])
iqr = q75 - q25

print("Range of Dataset:", range_value)
print("Interquartile Range (IQR):", iqr)
```

```python
#27 Implement Z-score normalization on a dataset and visualize its transformation


from scipy.stats import zscore
import matplotlib.pyplot as plt
import numpy as np

# Generate normal data
data = np.random.normal(50, 10, 1000)

# Perform Z-score normalization
normalized_data = zscore(data)

# Visualize original and normalized data
plt.figure(figsize=(10, 4))

# Original Data
plt.subplot(1, 2, 1)
plt.hist(data, bins=30)
plt.title('Original Data')

# Normalized Data
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30)
plt.title('Z-score Normalized Data')

plt.tight_layout()
plt.show()
```

```python
#28 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution


from scipy.stats import skew, kurtosis
import numpy as np

# Generate normal data
data = np.random.normal(0, 1, 1000)

# Calculate skewness and kurtosis
data_skewness = skew(data)
data_kurtosis = kurtosis(data)

print("Skewness of Dataset:", data_skewness)
print("Kurtosis of Dataset:", data_kurtosis)
```
